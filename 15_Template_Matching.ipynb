```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the full-body image and the head image (template)
full_body_image = cv2.imread('dhoni_template.jpg')
head_image = cv2.imread('dhoni.png')

# Convert images to grayscale (for template matching)
full_body_gray = cv2.cvtColor(full_body_image, cv2.COLOR_BGR2GRAY)
head_gray = cv2.cvtColor(head_image, cv2.COLOR_BGR2GRAY)

# Perform template matching
result = cv2.matchTemplate(full_body_gray, head_gray, cv2.TM_CCOEFF_NORMED)

# Get the best match location
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

# Define the rectangle around the matched area
top_left = max_loc
height, width = head_image.shape[:2]
bottom_right = (top_left[0] + width, top_left[1] + height)

# Draw a rectangle around the matched area on the full body image
cv2.rectangle(full_body_image, top_left, bottom_right, (0, 255, 0), 2)

plt.imshow(cv2.cvtColor(full_body_image, cv2.COLOR_BGR2RGB))
plt.show()
```
